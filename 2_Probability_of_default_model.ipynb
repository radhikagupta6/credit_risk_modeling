#### Independent variables
* <b> Fine-classing </b> - Turining continuous variables into categories. Through fine classing both discrete and continuous variables can be represented with categories.

* <b> Weight of Evidence </b>- To what extent an independent variable would predict a dependent variable. The farther away from zero the weight of evidence the better the category in differentiating between the outcomesof the dependent variable.

<b> WoE = ln(% good / % bad) </b>

* <b> Coarse classing </b> - The process of constructing new categories based on the initial ones. 


#### Information value
 - The extent to which an original independent variable, split into categories, is informative of an outcome variable.
 - Weight of evidence and the difference of the proportions of two outcomes across all categories of an original independent variable.
 - How much information the original independent variable brings with respect to explaining the dependent variable
 - information value can be used to identify the original independent variables which explain the dependent variable best.

Hence as you are probably anticipating you can use information value for pre-selection. So if we have 100 contestants for predictors in a regression the information value can help us preselect the best five or 10 or 20 independent variables to include in a statistical model the information value was always in the range from 0 to 1 greater values mean that the original variable yields more information with respect to the outcome.

#### Preprocessing Discrete Variables: Automating Calculations

* parameters are - independent variable dataframe, discrete local dependent variable

```python
def woe_discrete(df,discrete_variable_name, good_bad_variable_df):
    df = pd.concat([df[discrete_variable_name], good_bad_variable_df], axis = 1)
    df = pd.concat([df.groupby(df.columns.values[0], as_index= False)[df.columns.values[1]].count(),
                df.groupby(df.columns.values[0], as_index= False)[df.columns.values[1]].mean()], axis=1)
    # we keep the 1,2,4 columns
    df = df.iloc[:, [0,1,3]]
    # we assign the column names
    df.columns = [df.columns.values[0], 'n_obs', 'prop_good']
    # we calculate the proportion of observation from the total number of observation
    df['prop_n_obs'] = df['n_obs'] / df['n_obs'].sum()    
    # We calculate number of good and bad borrowers
    df['n_good'] = df['prop_good']* df['n_obs']
    df['n_bad'] = (1-df['prop_good'])* df['n_obs']
    # We calculate the proportions of good and bad borrowers from the total number of borrowers
    df['prop_n_good'] = df['n_good']/ df['n_good'].sum()
    df['prop_n_bad'] = df['n_bad']/ df['n_bad'].sum()
    # we calculate weight of evidence
    df['WoE'] = np.log(df['prop_n_good']/ df['prop_n_bad'])
    # we sort the values by weight of evidence and reset the index 
    df = df.sort_values(['WoE'])
    df = df.reset_index(drop = True)
    # we calculate the differences in proportions of good loans and weight of evidence 
    df['diff_prop_good'] = df['WoE'].diff().abs()
    df['diff_WoE'] = df['WoE'].diff().abs()
    # we find the information value
    df['IV'] = (df['prop_n_good'] - df['prop_n_bad']) * df['WoE']
    df['IV'] = df['IV'].sum()
    return df 
    
```

###### Function to visualize the data
* matplotlib does not fare well with the dfs and string but works fine with numpy arrays

```python
def plot_by_woe(df_WoE, rotation_of_x_axis_labels = 0):
    x = np.array(df_WoE.iloc[:,0].apply(str))
    y = df_WoE['WoE']
    # figsize = (width, height)
    plt.figure(figsize = (18,6))
    # market = 'o' displays a dot at each data point
    plt.plot(x,y, marker = 'o', linestyle = '--', color = 'k')
    plt.xlabel(df_WoE.columns[0])
    plt.ylabel('Wight of Evidence')
    plt.title(str('Weight of Evidence by' + df_WoE.columns[0]))
    plt.xticks(rotation = rotation_of_x_axis_labels)
    
    
```

#### Preprocessing Discrete Variables: Creating Dummy Variables, Part - 1

Based on the WoE, we must decide how to organize the original categories of the discrete variables into dummy variables for the PD model.

From a set of categorical variables that represent one original independent variable, we make a reference category the category with lowest WoE.

```python
df_temp = woe_discrete(df_inputs_prepr, 'home_ownership',df_targets_prepr )
df_temp
```

```python
plot_by_woe(df_temp)
```

According to the graph the categories 'OTHER' and 'NONE'are associated with the highest probability of default... There are very few loans associated with these categories in the table above. The WoE can not be calculated for the category 'ANY' because there's only one observation with this category.

It makes no sense to have dummy variables for these categories. 

But we can't neglect these as it can be loss of information. In such cases, we combine underrepresented categories that are similar to them. 

 

```python
df_inputs_prepr['home_ownership: RENT_OTHER_NONE_ANY'] = sum([df_inputs_prepr['home_ownership:RENT'], df_inputs_prepr['home_ownership:OTHER'], df_inputs_prepr['home_ownership:NONE'], df_inputs_prepr['home_ownership:ANY']])
```
